```python
import pandas as pd
import numpy as np

import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# load the csv file
df = pd.read_csv('bank.csv')

# display the first five observations
df.head()
```

```python
# Cecking if outliers are present
df.boxplot()
```

```python
for x in df.select_dtypes(np.number):
    sns.boxplot(x=df[x])
    plt.show()
```

```python
# Removing Outliers
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1

df = df[~((df < (q1 - (1.5 * IQR))) | (df > (q3 + (1.5 * IQR)))).any(axis=1)]
```

```python
df.boxplot()
```

```python
for x in df.select_dtypes(np.number):
    sns.boxplot(x=df[x])
    plt.show()
```
